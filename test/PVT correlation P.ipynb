# Вязкость
## Вязкость дегазированной нефти, корреляция Стендинга
$$
\mu_o = (0,32 + \frac{1,8 *10^7}{\gamma_{API}^{4,53}})(\frac{360}{1,8T-260})^a
$$
где:

$\gamma_{API} = \frac{141,5}{\gamma_o} - 131,5$ - плотность нефти в $ ^{\circ}\mathrm{API}$ 

$a = 10^{0,43 + 8,33 / \gamma_{API}} $

T - температура в Кельвинах

Графическая корреляция была установлена Билом. Математическую интерполяции корреляции Била дал Стендинг.

Источник: Дж.П. Брилл, Х.Мукеражди "Многофазный поток в скважинах"



```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from sympy import *
```

```python
#Функция рассчитывающая вязкость по Билу
def dead_oil_viscosity__Standing(gamma_o, t__C):
    gamma_API = 141.5 / gamma_o - 131.5
    viscosity = (0.32 + 1.8 * (10 ** 7) / (gamma_API ** 4.53)) * \
    (360 / (1.8 * (t__C + 273) - 260)) ** (10 ** (0.43 + 8.33 / (gamma_API)))
    return viscosity
```

```python
#код для построения графиков
t_set = np.array(range(40, 120, 20))
gamma_o_set = np.arange( 0.78, 1, 0.001)
plt.figure(figsize=(15,6))
#рисуем первый график
plt.subplot(121)
for t in t_set:
    visc__cP = dead_oil_viscosity__Standing(gamma_o_set, t)
    plt.plot([go*1000 for go in gamma_o_set], visc__cP, label='t=%1.0f$^{\circ}\mathrm{C}$'%t)
plt.legend()
plt.title('Вязкость дегазированной нефти от плотности')
plt.xlabel('Плотность, кг/м3')
plt.ylabel('Вязкость, сПз')
#рисуем второй график(для наглядности ограничиваем плотность дегазированной нефти)    
plt.subplot(122)
for t in t_set:
    visc__cP = dead_oil_viscosity__Standing(gamma_o_set, t)
    plt.plot([go*1000 for go in gamma_o_set], visc__cP, label='t=%1.0f$^{\circ}\mathrm{C}$'%t)
plt.xlim(775, 900)
plt.ylim(0,20)
plt.legend()
plt.title('Вязкость дегазированной нефти от плотности(775-900)')
plt.xlabel('Плотность, кг/м3')
plt.ylabel('Вязкость, сПз')
plt.show()
```
